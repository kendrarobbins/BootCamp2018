```python
import scipy.special
```

```python
scipy.special.binom??
```

```python
def prob2(x,v,n):
    
    A = scipy.special.binom(n, v) * x**v * (1 - x)**(n - v)
    return A
```

```python
x = np.linspace(0,1,100)
```

```python
b = [0,1,2,3]

for n in b:
    for v in np.arange(0,n+1,1):
        plt.plot(x, prob2(x,v,n))
        plt.title('n ='+ str(n)+',' 'v='+ str(v))
        plt.xlim(0,1)
        plt.ylim(0,1)
        plt.xlabel('x')
        plt.ylabel(r"$b_{v,n}(x)$")
        plt.show()
```
