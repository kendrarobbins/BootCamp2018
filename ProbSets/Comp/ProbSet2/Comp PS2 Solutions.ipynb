```python
def prob2(x,v,n):
    
    A = scipy.special.binom(n, v) * x**v * (1 - x)**(n - v)
    return A
```

```python
x = np.linspace(0,1,100)
```

```python
b = [0,1,2,3]

for n in b:
    for v in np.arange(0,n+1,1):
        plt.plot(x, prob2(x,v,n))
        plt.title('n ='+ str(n)+',' 'v='+ str(v))
        plt.xlim(0,1)
        plt.ylim(0,1)
        plt.xlabel('x')
        plt.ylabel(r"$b_{v,n}(x)$")
        plt.show()
```

```python
def prob2(p, d=100):
    
    i = pd.date_range("1/1/2000", "12/31/2000", freq='D')
    
    j = np.zeros(len(i))
    j = bernoulli.rvs(p, size=len(i))
    j[0]=d
    
    s = pd.Series(j, i, name = "Series")
   
    value = np.zeros(len(i))
    value = s.cumsum()
    
    plt.plot(i, value)
    plt.show
   
```

```python
setosa.plot()
plt.ylim(0,8)
plt.title("Setosas")
plt.ylabel("measurements in centimeters")
versicolor.plot()
plt.title("Versicolors")
plt.ylim(0,8)
plt.ylabel("measurements in centimeters")


virginica.plot()
plt.title("Virginicas")
plt.ylim(0,8)
plt.ylabel("measurements in centimeters")



plt.show()
```

Setosas are easiest to distinguish because their sepal width is larger than thier petal length, and petal width is very small.

to identify an iris:

    if petal width is less than one centimeter, it's probably a setosa.
    
    if petal length is roughly 4-5 cm, it's probably a versicolor.
    
    otherwise, it's probably a virginica.
