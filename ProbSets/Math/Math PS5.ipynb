# 8.1

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0,6,500)
y1 = (4+2*x)/3
y2 = (x-1)/6
y3 = 6-x


plt.figure(figsize=(10, 10))
plt.plot(x,y1,label = "constraint 1")
plt.plot(x,y2, label = "constraint 2")
plt.plot(x,y3, label = "constraint 3")
plt.xlim(0,6)
plt.fill_between(x, y3, y1, color='grey', where =  y3 >= y1)

#obj = (5x - c)/4
obj_1 = (5*x - 1) / 4 #objective function = 1
obj_2 = (5*x - 5) / 4 #objective function = 5
obj_3 = (5*x + 3) / 4 #objective function = -3

plt.plot(x, obj_1, '--', label = "objective function = 1")
plt.plot(x, obj_2, '--', label = "objective function = 5")
plt.plot(x, obj_3, '--', label = "objective function = -3")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

By solving for the appropriate vertex intersected by the red line, we see that the optimum occurs at x = 2.8 and y = 3.2. 

# 8.2 i)

```python
x1 = np.linspace(0,30,5000)

x2_1 = (15-x1)/3
x2_2 = (18-2*x1)/3
x2_3 = (x1-4)


plt.figure(figsize = (10,10))
plt.plot(x1,x2_1, label = "constraint 1")
plt.plot(x1, x2_2, label ="constraint 2")
plt.plot(x1, x2_3, label = "constraint 3")
plt.fill_between(x1, x2_3, x2_1, where = x2_2 >= x2_1, color = 'grey',interpolate = True)
plt.fill_between(x1, x2_3, x2_2, where = x2_2 <= x2_1, color='grey',interpolate = True)
plt.xlim(0,6)
plt.ylim(0,8)

obj_22 = 5-3*x1
obj_33 = 10-3*x1
obj_44 = 15 -3*x1

plt.plot(x1, obj_22, '--',label = "objective function = 5")
plt.plot(x1, obj_33, '--',label = "objective function = 10")
plt.plot(x1, obj_44, '--',label = "objective function = 15")
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()

plt.plot()
```

The vertex (6,2) is the optimizer.

```python
x = np.linspace(0,40,500)

y1 = x+11
y2 = 27-x
y3 = (90-2*x)/5

plt.figure(figsize =(10,10))
plt.plot(x,y1, label = "constraint 1")
plt.plot(x,y2,label = "constraint 2")
plt.plot(x,y3,label = "constraint 3")

plt.fill_between(x, 0, y1, color='grey', where = y3 >= y1, interpolate = True)
plt.fill_between(x, 0, y3, color='grey', where = (y3 <= y1)&(y3 <= y2), interpolate = True)
plt.fill_between(x, 0, y2, color='grey', where = y2 <= y3, interpolate = True)
plt.xlim(0,40)
plt.ylim(0,40)

obj_2 = (50-4*x)/6
obj_3 = (100-4*x)/6
obj_4 = (125-4*x)/6

plt.plot(x, obj_2, '--', label = "objective function =50")
plt.plot(x, obj_3, '--', label = "objective function = 100")
plt.plot(x, obj_4, '--', label = "objective function = 125")
plt.xlabel('x')
plt.legend()
plt.ylabel('y')

plt.plot()
```
